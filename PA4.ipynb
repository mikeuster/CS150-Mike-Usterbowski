## Task 1: Input Positive Integer and Factors


```python
num = int(input("Please input a positive integer: "))

for i in range(1, num + 1):
    if num % i == 0:
        print(i, "is a factor of", num)
        
```

```python
num = int(input("Please input a positive integer: "))

factor_count = 0

for i in range(1, num + 1):
    if num % i == 0:
        factor_count += 1

if factor_count == 2:
    print("Your number is prime")
else:
    print("Your number is not prime")
    
```

## Task 2: Language Translator

```python
translator = {
    "hello": "hola",
    "goodbye": "adios",
    "dog": "perro",
    "cat": "gato",
    "house": "casa",
    "water": "agua",
    "food": "comida",
    "chair": "silla",
    "book": "libro",
    "school": "escuela"
}

word = input("Enter a word to translate from English to Spanish (hit enter to stop): ").lower()

while word != "":
    if word in translator:
        print("The Spanish word for", word, "is", translator[word].upper() + ".")
    else:
        print("The Spanish word for", word, "is not in this dictionary.")
    
    word = input("Enter a word to translate from English to Spanish (hit enter to stop): ").lower()
```

## Task 3: Temperature Summaries

```python
temps = {
    "january": [43, 28],
    "february": [47, 31],
    "march": [58, 40],
    "april": [69, 49],
    "may": [78, 59],
    "june": [86, 67],
    "july": [87, 70],
    "august": [86, 69],
    "september": [80, 62],
    "october": [69, 50],
    "november": [57, 40],
    "december": [46, 32]
}

month = input("Enter a month: I will give you the average high and low temperatures in Louisville: ").lower()

while month != "quit":
    if month in temps:
        print("The average high temperature in Louisville for", month.capitalize(), "is", temps[month][0], "and the average low is", temps[month][1], ".")
    else:
        print("Not a valid month")
    
    month = input("Enter a month: I will give you the average high and low temperatures in Louisville: ").lower()
```

## Task 4: Inventory Dictionary and Prices

```python
inventory = {
    "hammer": [15, 12.49],
    "wrench": [20, 7.99],
    "screwdriver": [40, 3.99],
    "nailgun": [10, 36.59],
    "padlock": [25, 5.99],
    "roll_of_duct_tape": [30, 4.95]
}

choice = ""

while choice != "5":
    print("\nMenu")
    print("1. Search for an item")
    print("2. Show items under a budget")
    print("3. Show total quantity of items")
    print("4. Show total price of all inventory")
    print("5. Quit")
    
    choice = input("Enter your choice: ")
    
    if choice == "1":
        item = input("Enter an item name: ").lower()
        
        if item in inventory:
            print("There are", inventory[item][0], item + "(s) available at $" + str(inventory[item][1]), "each.")
        else:
            print("That item is not available.")
    
    elif choice == "2":
        budget = float(input("Enter a maximum budget amount: "))
        print("Items below your budget:")
        
        for item in inventory:
            if inventory[item][1] < budget:
                print(item, "- $" + str(inventory[item][1]))
    
    elif choice == "3":
        total_quantity = 0
        
        for item in inventory:
            total_quantity += inventory[item][0]
        
        print("The total quantity of items available is", total_quantity)
    
    elif choice == "4":
        total_price = 0
        
        for item in inventory:
            total_price += inventory[item][0] * inventory[item][1]
        
        print("The total price of all items across the inventory is $" + str(round(total_price, 2)))
    
    elif choice == "5":
        print("Goodbye")
    
    else:
        print("Invalid menu choice")
        
```
