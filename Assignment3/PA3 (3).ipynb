```python
# Task 1: Types of Triangles

side1 = float(input("Enter side 1: "))
side2 = float(input("Enter side 2: "))
side3 = float(input("Enter side 3: "))

if side1 == side2 == side3:
    print("This is an equilateral triangle.")
elif side1 == side2 or side1 == side3 or side2 == side3:
    print("This is an isosceles triangle.")
else:
    print("This is a scalene triangle.")
```

```python
# Task 2: Student Grades and At-Risk Status

student_grades = []

for i in range(5):
    grade = float(input("Enter grade average for student " + str(i + 1) + ": "))
    student_grades.append(grade)

class_average = sum(student_grades) / len(student_grades)
min_grade = min(student_grades)
max_grade = max(student_grades)

print("Class average:", class_average)
print("Minimum grade:", min_grade)
print("Maximum grade:", max_grade)

at_risk_count = 0
for grade in student_grades:
    if grade < 65:
        at_risk_count += 1

print(str(at_risk_count) + " students are at risk in the course.")
```

```python
# Task 3: Ranges and Multiples

mult_of_2 = list(range(2, 101, 2))
mult_of_3 = list(range(3, 101, 3))

mult_of_both = []
for num in mult_of_2:
    if num in mult_of_3:
        mult_of_both.append(num)

mult_of_either = []
for num in range(1, 101):
    if num in mult_of_2 or num in mult_of_3:
        mult_of_either.append(num)

print("Multiples of 2:", mult_of_2)
print("Multiples of 3:", mult_of_3)
print("Multiples of both 2 and 3:", mult_of_both)
print("Multiples of either 2 or 3:", mult_of_either)
```

```python
# Task 4: Shopping List

shopping_list = []

for i in range(7):
    item = input("Enter an item for your shopping list (or type 'done' to stop): ")
    
    if item.lower() == "done":
        break
    
    shopping_list.append(item)

print("Total number of items on the shopping list:", len(shopping_list))

print("Shopping list in reverse order:")
for item in shopping_list[::-1]:
    print(item)

print("Shopping list in alphabetical order:")
for item in sorted(shopping_list):
    print(item)

duplicate_found = False

for i in range(len(shopping_list)):
    for j in range(i + 1, len(shopping_list)):
        if shopping_list[i].lower() == shopping_list[j].lower():
            duplicate_found = True

if duplicate_found:
    print("At least one item appears more than once in the shopping list.")
else:
    print("No items appear more than once in the shopping list.")
```
