## Task 1: Weekend Activities as Sets

```python
# Task 1: Weekend Activities as Sets

# Saturday activities
actSat = set()

while True:
    activity = input('Enter an activity from last Saturday ("Done" to stop): ')
    if activity == "Done":
        break
    actSat.add(activity)

# Sunday activities
actSun = set()

while True:
    activity = input('Enter an activity from last Sunday ("Done" to stop): ')
    if activity == "Done":
        break
    actSun.add(activity)

# Print both sets
print("Saturday activities:", actSat)
print("Sunday activities:", actSun)

# Activities done on both days
print("Activities done on both Saturday and Sunday:",
      actSat.intersection(actSun))

# Activities done on at least one day
print("Activities done on at least one of the days:",
      actSat.union(actSun))
```

## Task 2: Triangle Calculations


```python
# Task 2: Triangle Calculations

import math

def triangle_solve(leg1, leg2):
    
    # Calculate hypotenuse
    hypotenuse = math.sqrt(leg1**2 + leg2**2)
    
    # Calculate perimeter
    perimeter = leg1 + leg2 + hypotenuse
    
    # Calculate area
    area = 0.5 * leg1 * leg2
    
    # Return results as a tuple
    return (hypotenuse, perimeter, area)


# Test the function (recommended by your professor)
result = triangle_solve(3, 4)

print("Result:", result)
```

## Task 3: Calculator Class


```python
# Task 3: Calculator Class

class Calculator:
    
    def __init__(self):
        self.prev_result = None
        self.total_calculations = 0

    def add(self, a, b):
        result = a + b
        self.prev_result = result
        self.total_calculations += 1
        return result

    def subtract(self, a, b):
        result = a - b
        self.prev_result = result
        self.total_calculations += 1
        return result

    def multiply(self, a, b):
        result = a * b
        self.prev_result = result
        self.total_calculations += 1
        return result

    def divide(self, a, b):
        result = a / b
        self.prev_result = result
        self.total_calculations += 1
        return result

    def power(self, a, b):
        result = a ** b
        self.prev_result = result
        self.total_calculations += 1
        return result

    def info(self):
        print("Previous result:", self.prev_result)
        print("Total calculations:", self.total_calculations)


# Testing the Calculator (recommended)
calc = Calculator()

print("Add:", calc.add(10, 5))
print("Subtract:", calc.subtract(10, 5))
print("Multiply:", calc.multiply(10, 5))
print("Divide:", calc.divide(10, 5))
print("Power:", calc.power(2, 3))

calc.info()
```
