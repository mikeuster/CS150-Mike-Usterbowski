## Task 2: Distance Conversion
Convert a distance in miles to kilometers and print both values.

```python
miles_distance = 21
kilometers = miles_distance * 1.60934

print(f"The distance is {miles_distance} miles, which is the equivalent of {kilometers:.2f} kilometers.")
```

## Task 3: Working with Running Times
Use two running times (in minutes) and perform several arithmetic operations.

```python

run1 = 26.7
run2 = 29.1
total_time = run1 + run2
improvement = run1 - run2
twice_run2 = run2 * 2
average_time = (run1 + run2) / 2
print(f"Run 1 time: {run1} minutes")
print(f"Run 2 time: {run2} minutes")
print(f"Total time spent running: {total_time} minutes")
print(f"Improvement (run1 - run2): {improvement} minutes")
print(f"Run 2 done twice: {twice_run2} minutes")
print(f"Average of both runs: {average_time} minutes")
```

## Task 4: String Manipulation
Work with a famous quote and perform several string operations.


```python
quote = "  hard work beats talent when talent fails to work hard.   "

print("Quote:", quote)
print("Length of quote:", len(quote))
print("Uppercase:", quote.upper())
print("Lowercase:", quote.lower())
print("Stripped (no leading/trailing whitespace):", quote.strip())
print("Split into words:", quote.strip().split())
```

## Task 5: User Input and Operations
Prompt the user for two leg lengths of a right triangle and compute the hypotenuse using the Pythagorean theorem.

```python
import math

a = float(input("Enter the length of leg a:3 "))
b = float(input("Enter the length of leg b:5 "))

c = math.sqrt(a**2 + b**2)

print(f"The hypotenuse of your triangle is {c:.2f} units.")


```
