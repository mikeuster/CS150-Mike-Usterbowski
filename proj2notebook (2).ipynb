```python
import pandas as pd

# Load dataset directly from link
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv"

df = pd.read_csv(url)

# Show first rows
df.head()
```

```python
category_col = "day"
num_col1 = "total_bill"
num_col2 = "tip"

# Overall statistics for total_bill
print("Total Bill Statistics")
print("Mean:", df[num_col1].mean())
print("Max:", df[num_col1].max())
print("Min:", df[num_col1].min())
print("Std Dev:", df[num_col1].std())
print("Range:", df[num_col1].max() - df[num_col1].min())
print("Median:", df[num_col1].median())

print()

# Overall statistics for tip
print("Tip Statistics")
print("Mean:", df[num_col2].mean())
print("Max:", df[num_col2].max())
print("Min:", df[num_col2].min())
print("Std Dev:", df[num_col2].std())
print("Range:", df[num_col2].max() - df[num_col2].min())
print("Median:", df[num_col2].median())
```

```python
# Statistics by category (day)

grouped = df.groupby(category_col)

stats = grouped[[num_col1, num_col2]].agg(
    ['mean', 'max', 'min', 'std', 'median']
)

print(stats)
```

```python
import matplotlib.pyplot as plt

# Scatter plot
plt.scatter(df[num_col1], df[num_col2])

plt.xlabel("Total Bill")
plt.ylabel("Tip")
plt.title("Scatter Plot of Total Bill vs Tip")

plt.show()
```

```python
# Histogram for total_bill

df[num_col1].hist()

plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill")
plt.ylabel("Frequency")

plt.show()
```

```python
# Histogram for tip

df[num_col2].hist()

plt.title("Histogram of Tip")
plt.xlabel("Tip")
plt.ylabel("Frequency")

plt.show()
```

```python
# Histogram by category (day)

categories = df[category_col].unique()

for day in categories:
    df[df[category_col] == day][num_col1].hist()

    plt.title(f"Total Bill Distribution for {day}")
    plt.xlabel("Total Bill")
    plt.ylabel("Frequency")

    plt.show()
```
